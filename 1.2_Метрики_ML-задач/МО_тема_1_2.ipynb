#1. Бинаризуйте переменную «Статус»

```python
# Вывод количества уникальных элементов столбца 'Статус'
df['Статус'].value_counts()
```

Статусы:

-1 - студент находится в академическом отпуске

 1 - учится в университете

 3 - отчислен

 4 - закончил обучение

 6 - архив

 **Итог: 0 – отчислится, 1 – не отчислится**

```python
# Словарь
status = {-1: 1, 1: 1, 3: 0, 4: 1, 6: 0}
df['Статус'] = df['Статус'].replace(status)

df.head(10)
```

#7. Придумайте способы кодирования имеющихся признаков

```python
# Признак 'Основание' можно преобразовать в числовой формат
mapping = {'СН': 1, 'ОО': 2, 'ДН': 3, 'БН': 4, 'ЦН': 5, 'ЛН': 6}
df['Основания'] = df['Основания'].map(mapping)
df['Основания']
```

```python
# Признак 'Тип_УЗ' можно объединить в 5 групп:
# 1 [школа, школа-интернат, сош]
# 2 [гимназия, лицей, школа-лицей]
# 3 [колледж, техникум, училище, пту]
# 4 [университет, академия, институт]
# 5 [учреждение, другое, нпо]
```

```python
df['Тип_УЗ'].value_counts()
```

```python
emb = {"2017": 'школа'} # 'школа' - часто встречающееся значение
df['Тип_УЗ'] = df['Тип_УЗ'].replace(emb)
```

```python
# Словарь меток групп
group_dict = {
    'школа': 1,
    'школа-интернат': 1,
    'сош': 1,
    'гимназия': 2,
    'лицей': 2,
    'школа-лицей': 2,
    'школа-гимназия': 2,
    'колледж': 3,
    'техникум': 3,
    'училище': 3,
    'пту':3,
    'университет': 4,
    'академия': 4,
    'институт': 4,
    'другое': 5,
    'нпо': 5,
    'учреждение': 5}


df['Тип_УЗ'] = df['Тип_УЗ'].map(group_dict)
```

```python
df['Тип_УЗ'].value_counts()
```

# 8. Придумайте способы генерации новых признаков

```python
# Новый признак 1: иностранный гражданин как бинаризация 'Страна_ПП'
#df['иностранный гражданин'] = 9
for i in range(len(df['Страна_ПП'])):
  if (df.loc[i, 'Страна_ПП'] != 'россия'):
    df.loc[i, 'иностранный гражданин'] = 1
  else:
    df.loc[i, 'иностранный гражданин'] = 0

df['иностранный гражданин']

# Полученный значения
df['иностранный гражданин'].value_counts()
```
