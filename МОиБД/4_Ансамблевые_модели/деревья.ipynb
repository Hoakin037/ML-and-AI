## <center> Регрессионные модели, основанные на деревьях решений

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

red_df = pd.read_csv("winequality-red.csv", sep=";", decimal=".", encoding="ansi")
white_df = pd.read_csv("winequality-white.csv", sep=";", decimal=".", encoding="ansi")
```

```python
#  Ваш код здесь
y_red = red_df['quality']
y_red = y_red.values
X_red = red_df[['alcohol', 'sulphates', 'volatile acidity', 'citric acid']]

y_white = white_df['quality']
y_white = y_white.values
X_white = white_df[['alcohol', 'sulphates', 'volatile acidity', 'citric acid']]
```

```python
#  Ваш код здесь
from sklearn.model_selection import train_test_split

X_white_train, X_white_test, y_white_train, y_white_test = train_test_split(X_white, y_white, random_state=0, test_size=0.3)

X_red_train, X_red_test, y_red_train, y_red_test = train_test_split(X_red, y_red, random_state=0, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

model=DecisionTreeRegressor(random_state=42,max_depth=5)
model.fit(X_red_train, y_red_train)
y_red_pred = model.predict(X_red_test)
for i in range (len(y_red_pred)): print("i=", i," | y_pred=", y_red_pred[i]," | y_test=", y_red_test[i])
```

```python
model=DecisionTreeRegressor(random_state=42,max_depth=5)
model.fit(X_white_train, y_white_train)
y_white_pred = model.predict(X_white_test)
for i in range (len(y_white_pred)): print("i=", i," | y_pred=", y_white_pred[i]," | y_test=", y_white_test[i])
```

```python
print ('RMSE красное=', np.sqrt(mean_squared_error(y_red_pred,y_red_test)))
print ('RMSE белое =', np.sqrt(mean_squared_error(y_white_pred,y_white_test)))
```
