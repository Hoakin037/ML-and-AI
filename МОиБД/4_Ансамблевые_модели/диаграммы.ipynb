Красное

```python
import pandas as pd
import matplotlib.pyplot as plt

red_df = pd.read_csv("winequality-red.csv", sep=";", decimal=".", encoding="ansi")

plt.ylabel('quality')
plt.xlabel('alcohol')
plt.scatter(x=red_df['alcohol'], y=red_df['quality'])
```

```python
plt.figure(figsize=[10,5])
plt.ylabel('quality')
plt.xlabel('volatile acidity')
plt.scatter(x=red_df['volatile acidity'], y=red_df['quality'])
```

```python

plt.ylabel('quality')
plt.xlabel('citric acid')
plt.scatter(x=red_df['citric acid'], y=red_df['quality'])
```

```python
plt.ylabel('quality')
plt.xlabel('sulphates')
plt.scatter(x=red_df['sulphates'], y=red_df['quality'])
```
