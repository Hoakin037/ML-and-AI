```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.linear_model import LinearRegression as lr
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split


red_df = pd.read_csv("winequality-red.csv", sep=";", decimal=".", encoding="ansi")
white_df = pd.read_csv("winequality-white.csv", sep=";", decimal=".", encoding="ansi")
```

```python
plt.figure(figsize=(8,8), dpi=90)
sn.heatmap(red_df.corr(), xticklabels=red_df.corr().columns, yticklabels=red_df.corr().columns, annot=True)
```

```python
plt.figure(figsize=(10,10), dpi=90)
sn.heatmap(white_df.corr(), xticklabels=white_df.corr().columns, yticklabels=white_df.corr().columns, annot=True)
```

```python
#    
y_red = red_df['quality']
y_red = y_red.values
X_red = red_df[['alcohol', 'sulphates', 'volatile acidity', 'citric acid']]

y_white = white_df['quality']
y_white = y_white.values
X_white = white_df[['alcohol', 'sulphates', 'volatile acidity', 'citric acid']]
```

```python
#    
X_white_train, X_white_test, y_white_train, y_white_test = train_test_split(X_white, y_white, random_state=0, test_size=0.3)

X_red_train, X_red_test, y_red_train, y_red_test = train_test_split(X_red, y_red, random_state=0, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression as lr
model_white = lr(max_iter=1000)
model_white.fit(X_white_train, y_white_train)
model_white.coef_
```

```python
model_red = lr(max_iter=1000)
model_red.fit(X_red_train, y_red_train)
model_red.coef_
```

```python
#    
y_red_pred = model_red.predict(X_red_test)
for i in range(len(y_red_pred)): print('i=', i, '|  =', y_red_test[i],'|  =', y_red_pred[i])
pd.crosstab(y_red_test, y_red_pred)
```

```python
#    
y_white_pred = model_white.predict(X_white_test)
for i in range(len(y_white_pred)): print('i=', i, '|  =', y_white_test[i],'|  =', y_white_pred[i])
pd.crosstab(y_white_test, y_white_pred)
```

```python
#    
rmse_red = np.sqrt(mean_squared_error(y_red_test, y_red_pred))
print('RMSE_test () = {:.3f}'.format(rmse_red))
rmse_white = np.sqrt(mean_squared_error(y_white_test, y_white_pred))
print('RMSE_test () = {:.3f}'.format(rmse_white))
```
