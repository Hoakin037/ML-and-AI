```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

y = df['Weight'].map(species_to_id)

X = df.drop('Weight', axis=1)

scaler = MinMaxScaler()
X_normalized = pd.DataFrame(
    scaler.fit_transform(X),
    columns=X.columns
)

df_shuffled = pd.concat([X_normalized, y], axis=1).sample(frac=1, random_state=42).reset_index(drop=True)

y = df['Weight']
X = df.drop('Weight', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
model.coef_
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Функция для расчёта MAPE (Mean Absolute Percentage Error)
# Добавляем небольшую константу, чтобы избежать деления на ноль
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    # Избегаем деления на ноль
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

# Расчёт метрик для обучающей выборки
mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)
mae_train = mean_absolute_error(y_train, y_train_pred)
mape_train = mean_absolute_percentage_error(y_train, y_train_pred)

# Расчёт метрик для тестовой выборки
mse_test = mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)
mape_test = mean_absolute_percentage_error(y_test, y_test_pred)

# Вывод результатов в красивой таблице
print("=== Метрики качества модели LinearRegression===")
print(f"{'Метрика':<15} {'Обучающая':<15} {'Тестовая':<15}")
print("-" * 45)
print(f"{'MSE':<15} {mse_train:<15.4f} {mse_test:<15.4f}")
print(f"{'R²':<15} {r2_train:<15.4f} {r2_test:<15.4f}")
print(f"{'MAE':<15} {mae_train:<15.4f} {mae_test:<15.4f}")
print(f"{'MAPE (%)':<15} {mape_train:<15.2f} {mape_test:<15.2f}")
print(f"{'MAE':<15} {mae_train:<15.4f} {mae_test:<15.4f}")
print(f"{'MAPE (%)':<15} {mape_train:<15.2f} {mape_test:<15.2f}")
```

```python
from sklearn.preprocessing import StandardScaler

def print_metrics(y_true, y_pred, name):
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    # MAPE с защитой от деления на ноль
    mape = np.mean(np.abs((y_true - y_pred) / np.where(y_true == 0, 1e-8, y_true))) * 100
    print(f"{name}")
    print(f"   MSE:  {mse:10.2f} | R²: {r2:.4f} | MAE: {mae:8.2f} | MAPE: {mape:.2f}%")
    print("-" * 60)


df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)

y = df_shuffled['Weight']
X = df_shuffled.drop('Weight', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)


std_scaler = StandardScaler()
X_train_std = std_scaler.fit_transform(X_train)
X_test_std = std_scaler.transform(X_test)

model_std = LinearRegression()
model_std.fit(X_train_std, y_train)

y_train_pred_std = model_std.predict(X_train_std)
y_test_pred_std = model_std.predict(X_test_std)

print_metrics(y_train, y_train_pred_std, "3. StandardScaler (train)     ")
print_metrics(y_test,  y_test_pred_std,  "3. StandardScaler (test)      ")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

y = df_shuffled['Species']
X = df_shuffled.drop('Species', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

results = []
for depth in range(1, 15):
    model = DecisionTreeClassifier(max_depth=depth, random_state=42)
    model.fit(X_train, y_train)
    
    train_acc = accuracy_score(y_train, model.predict(X_train))
    test_acc = accuracy_score(y_test, model.predict(X_test))
    
    results.append({'max_depth': depth, 'train_acc': train_acc, 'test_acc': test_acc})

# Таблица результатов
results_df = pd.DataFrame(results)
print(results_df)
```

Правило: самое маленькое значение глубины, при котором достигается максимальная test_accuracy.
Это даёт наиболее простую и устойчивую модель.

```python
model = DecisionTreeClassifier(max_depth=9, random_state=0)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_train_pred = model.predict(X_train).astype(int)
y_test_pred = model.predict(X_test).astype(int)

print("Тип y_train:", y_train.dtype)
print("Тип y_train_pred:", y_train_pred.dtype)
print("Тип y_test_pred:", y_test_pred.dtype)

# Метрики
print("\n=== Метрики качества (DecisionTree, max_depth=9) ===")
print(f"Train Accuracy:  {accuracy_score(y_train, y_train_pred):.4f}")
print(f"Test Accuracy:   {accuracy_score(y_test, y_test_pred):.4f}")

print(f"Train Precision (weighted): {precision_score(y_train, y_train_pred, average='weighted'):.4f}")
print(f"Test Precision (weighted):  {precision_score(y_test, y_test_pred, average='weighted'):.4f}")

print(f"Train Recall (weighted):    {recall_score(y_train, y_train_pred, average='weighted'):.4f}")
print(f"Test Recall (weighted):     {recall_score(y_test, y_test_pred, average='weighted'):.4f}")

print(f"Train F1-score (weighted):  {f1_score(y_train, y_train_pred, average='weighted'):.4f}")
print(f"Test F1-score (weighted):   {f1_score(y_test, y_test_pred, average='weighted'):.4f}")
```
